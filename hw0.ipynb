#Pandas

```python
#b c
import pandas as pd
data = pd.read_csv('Salaries.csv',skiprows=[1])#,index_col='playerID',header = 0
data = data.set_index('playerID')
data.head()
```

```python
#d
d = data.loc[(data.teamID.isin(['ATL','HOU'])) & (data.salary > 1000000)]
selected_id = d.index
selected_id
```

```python
#e
team_ATL = data.loc[data.teamID == 'ATL']
team_ATL_summary = team_ATL['salary'].describe()
team_ATL_summary
```

```python
#f
dict = {}
for index,row in data.iterrows():
  for column_name,value in row.items():
    if column_name not in dict:
      dict[column_name] = []
    dict[column_name].append(value)
print(dict)
```

```python
# to_dict method
data_dict2= data.to_dict()
print(data_dict2)
```

```python
#b
print(nparray.ndim)
print(nparray.shape)
print(nparray.size)
print(nparray.dtype)
print(nparray.itemsize)
print(nparray.data)
```
